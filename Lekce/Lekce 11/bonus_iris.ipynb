# `json`

```python
import os
import json


species = {}
with open(os.path.join("data", "iris.json")) as f:
    data = json.load(f)
    
for flower in data:  # data: List[Dict] => flower: Dict
    if flower["species"] not in species:
        species[flower["species"]] = {"sepalLength": [], "sepalWidth": [], "petalLength": [], "petalWidth": []}

    species[flower["species"]]["sepalLength"].append(flower["sepalLength"])
    species[flower["species"]]["sepalWidth"].append(flower["sepalWidth"])
    species[flower["species"]]["petalLength"].append(flower["petalLength"])
    species[flower["species"]]["petalWidth"].append(flower["petalWidth"])

for type_species in species.keys():
    avg_sepalLength = sum(species[type_species]["sepalLength"]) / len(
        species[type_species]["sepalLength"])
    avg_sepalWidth = sum(species[type_species]["sepalWidth"]) / len(
        species[type_species]["sepalWidth"])
    avg_petalLength = sum(species[type_species]["petalLength"]) / len(
        species[type_species]["petalLength"])
    avg_petalWidth = sum(species[type_species]["petalWidth"]) / len(
        species[type_species]["petalWidth"])

print(avg_sepalLength, avg_sepalWidth, avg_petalLength, avg_petalWidth)
```

```python
for type_species in species.keys():
    print(type_species)
    plt.scatter(
        species[type_species]["sepalWidth"], 
        species[type_species]["sepalLength"], 
        label=type_species
    )
plt.ylabel("sepalLength")
plt.xlabel("sepalWidth")
plt.legend();
```

```python
for type_species in species.keys():
    plt.scatter(species[type_species]["petalWidth"], species[type_species]["petalLength"], label=type_species)
plt.ylabel("petalLength")
plt.xlabel("petalWidth")
plt.legend();
```

```python
fig, axs = plt.subplots(1, 4, sharex=True, figsize=(15, 3))
for i, feature_name in enumerate(["sepalWidth", "sepalLength", "petalWidth", "petalLength"]):
    for type_species in species.keys():
        axs[i].hist(species[type_species][feature_name], label=type_species)
    axs[i].set_title(feature_name)
axs[0].legend();
```

```python
fig, axs = plt.subplots(1, 4, sharex=True, figsize=(15, 3))
for i, feature_name in enumerate(["sepalWidth", "sepalLength", "petalWidth", "petalLength"]):
    for j, type_species in enumerate(species.keys()):
        axs[i].boxplot(species[type_species][feature_name], positions=[j])
    axs[i].set_xticks(list(range(len(species))))
    axs[i].set_xticklabels(list(species.keys()))
    axs[i].set_title(feature_name)
```
